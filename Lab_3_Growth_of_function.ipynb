# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.



def binarySearch(array, target):
    left, right = 0, len(array) - 1
    steps = 0
    while left <= right:
        steps += 1
        mid = (left + right) // 2
        if array[mid] == target:
            return mid, steps
        elif array[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return -1, steps


sorted_array = [10, 100, 1000, 5000, 10000, 50000, 100000]
step_values = []

for n in sorted_array:
    data = list(range(n))
    _, steps = binarySearch(data, n-1)
    step_values.append(steps)

plt.figure()
plt.plot(sorted_array, step_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Steps taken")
plt.title("Binary Search Complexity ~ O(log n)")
plt.grid(True)
plt.show()


```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.d

def merge(arr, left, mid, right):

    n1 = mid - left + 1
    n2 = right - mid

    L = [0] * n1
    R = [0] * n2

    for i in range(n1):
        L[i] = arr[left + i]

    for j in range(n2):
        R[j] = arr[mid + 1 + j]

    i = j = 0
    k = left

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1


def mergeSort(arr, left, right):

    if left < right:
        mid = (left + right) // 2

        mergeSort(arr, left, mid)
        mergeSort(arr, mid + 1, right)
        merge(arr, left, mid, right)



arr = [38, 27, 43, 10]

mergeSort(arr, 0, len(arr) - 1)

print(arr)

```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in polynomial time. plot the graph as well, similar to above example

import matplotlib.pyplot as plt

def polynomial_time_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

n_values = list(range(1, 101, 2))
time_values = []

for n in n_values:
    steps = polynomial_time_example(n)
    time_values.append(steps)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', label='Actual O(n^2) Operations')

theoretical_n_squared = [n**2 for n in n_values]
plt.plot(n_values, theoretical_n_squared, linestyle='--', color='red', label='Theoretical O(n^2)')

plt.xlabel("Input size n")
plt.ylabel("Number of Operations")
plt.title("O(n^2) Polynomial Time Complexity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Log-Linear Time Complexity O(n log n) (average case)

import matplotlib.pyplot as plt
import numpy as np
import math
import random

def partition(arr, low, high, comparisons):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        comparisons[0] += 1
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quickSort_and_count_steps(arr, low, high, comparisons):
    if low < high:
        pi = partition(arr, low, high, comparisons)
        quickSort_and_count_steps(arr, low, pi - 1, comparisons)
        quickSort_and_count_steps(arr, pi + 1, high, comparisons)


n_values = list(range(1, 101, 5))
comparison_counts = []

for n in n_values:
    if n == 0:
        comparison_counts.append(0)
        continue

    data = list(range(n))
    random.shuffle(data)
    current_arr = list(data)

    comparisons = [0]
    quickSort_and_count_steps(current_arr, 0, len(current_arr) - 1, comparisons)
    comparison_counts.append(comparisons[0])

plt.figure(figsize=(10, 6))
plt.plot(n_values, comparison_counts, marker='o', linestyle='-', label='Actual Quick Sort Comparisons')

theoretical_n_log_n = [x * math.log2(x) if x > 1 else 0 for x in n_values]
plt.plot(n_values, theoretical_n_log_n, linestyle='--', color='red', label='Theoretical O(n log n)')

plt.xlabel("Input size n")
plt.ylabel("Number of Comparisons")
plt.title("Quick Sort Time Complexity: O(n log n) (Average Case)")
plt.legend()
plt.grid(True)
plt.show()
```
